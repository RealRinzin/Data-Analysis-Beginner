```python
df['fire_month'] = df.discovery_date.dt.month
df.head(10)
```

```python
# Using the date method to find total total days between contain_date and discovery_date
df['days_burning'] = (df.contain_date - df.discovery_date).dt.days
df.head(5)
```

```python
# Fire name change to the tile case ()
# change the fire name column to String Type
df['fire_name'] = df['fire_name'].str.title()
df.head(3)
```

```python
#Add Column derived from string Data
# Example : The Power Fire (2024) - The 'fire_name Fire ('fire_year')
df['full_name'] = 'The '+df['fire_name'] +' Fire ('+df['fire_year'].astype(str)+')'
df.head(2)
```

```python
#Add Column derived from numeric Computation
# such as 
df['acres_per_day'] = round(df.dropna().acres_burned / df.dropna().days_burning,2)
df.head(5)
```

##### Transform() Method

```python
# use Groupby method to get the mean value 
df['mean_day'] = round(df.groupby('state',observed=False)['days_burning'].transform('mean'),2)
df['acres_per_day'].isnull().sum()
df.head(10)
```

##### User defined method Apply() method

```python
# User defined functions
def convert_gender(data):
    if data.sex ==1:
        return 'male'
    elif data.sex == 2:
        return 'female'
    else:
        return 'non-binary'
# Applying the function to the data
works['sex'] = works.apply(convert_gender,axis=1)
works.head(5)
```

##### Lamda functions() method

```python
# def add(x,y):
#     return x+y
# print(add(4,5))
# (lambda x,y:x+y )(4,5)

# Working with inline lamda functions
works['sex'] = works.apply(lambda x: '1' if (x.sex =='male') else '0',axis=1)
# Transpose ROW and COLUMN
works.transpose().to_csv('transpose.csv',index=True)
# works.head(2)
```

```python
#File Input
exercise = pd.read_pickle('../Phase - 3/fires_by_month.pkl')
exercise.head(5)
```

```python
# Add column for the mean
exercise['acres_burned_per_day'] = round(exercise.acres_burned / exercise.days_burning,2)
exercise.head(4)
```

```python
# Add a column name with Lamda using If else
exercise['lamda'] = exercise.apply(
    lambda x : 0 if x.days_burning <=0 else round(x.acres_burned / x.days_burning,2), axis =1
)
exercise.head()
```

```python
# Convert the month to string 
# such as 0 = Jan, 1 = Feb

def months(row):
    month_dict = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr',
                  5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 
                  9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
    return month_dict[row.fire_month]

exercise['fire_month'] = exercise.apply(
    lambda x : months(x), axis=1
)
exercise.head(5)
```

```python
# set index on state, fire_year, fire_month

exercise = exercise.set_index(['state','fire_year','fire_month'])
exercise.head(4)
```
