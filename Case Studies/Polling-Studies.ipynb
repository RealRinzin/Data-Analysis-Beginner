```python
#import pandas
import pandas as pd
```

```python
#import the poll csv file
polls = pd.read_csv('./president_general_polls_2016.csv')
polls.info()
```

```python
polls.nunique()
```

```python
# Drop columns with only one unique value
polls.drop(columns=['cycle','branch','matchup','forecastdate'],inplace=True)
# keep now-cast type
polls = polls.query('type == "now-cast"')
polls.drop(columns='type',inplace=True)
polls.info()
```

```python
#Keep only rows with lv or rv in population column
polls.query('population !="Lv" & population !="rv"')
polls = polls.query('population in ["Lv","rv"]')
polls
```
