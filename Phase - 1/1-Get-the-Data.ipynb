### 5 Phases of Data Analysis
#### 1 - Get the Data

```python
import pandas as pd
```

```python
# Read the CSV file from local computer
df = pd.read_csv('../Data-Analysis-Beginner/CSV Files/data.csv') 
# Drop the duplicates
df.drop_duplicates(inplace=True)
# Drop Na rows
df.dropna(inplace=True)
# Remove the Calories Column from the records
df.drop('Calories',axis=1,inplace=True)
df.head(20)
```

```python
# SHAPE function
# get the total numbers of rows and total numbers of columns
df.shape
```

```python
# COLUMN Function
# Display all the column name of the record
df.columns
```

```python
# INFO() Function
# Check the record information of the CSV file such as
# total number of rows, total number of null, 
# Data types of each columns
# Total number of null values in each column
# Also memory Usage of the Record
df.info()
```

```python
# HEAD() function
# Display the first 8 records with the column name
df.head(8)
```

```python
# TAIL() function
# Display the last 19 records with the column name
df.tail(6)
```

```python
# DESCRIBE() function
# Get descriptive statistics about the record from each column
# such as COUNT,MEAN,MIN,STD,MAX 

df.describe()
```

```python
# SUBSET
# Display specific columns
df[['Duration','Pulse']].head(3) # we are trying to ge the first 3 rows from the record
```

```python
# QUERY Function
# Query or filter data from the record such as 
# Equal to , Greater Than, Less Than, AND , OR etc
df.query('(Duration == 60) and (Pulse > 130)')
```

```python
# SORT_VALUES
# Sort the rows with different orders such as ASC, DES order
df[['Calories']].sort_values('Calories',ascending=True).info
```

```python
df.apply(pd.unique)
```

```python
# To check the duplicate data in row
df.loc[df.duplicated()]
```
