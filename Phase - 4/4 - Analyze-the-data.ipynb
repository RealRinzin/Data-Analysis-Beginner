#### Adding the car csv files for the analyze  

```python
# importing pandas file and car csv files
import pandas as pd
# importing seaborn package
import seaborn as sns
#import file
cars = pd.read_csv('../Phase - 4/cars.csv')

cars[['aspiration','carbody','enginesize','curbweight','price']]
```

```python
# Using the Melt() function to melt the columns
melt_cars = pd.melt(cars, id_vars='price',
                    value_vars=['enginesize','curbweight'],
                    var_name='feature',
                    value_name='featureValue',
                    )
melt_cars
```

```python
# plotting with the seaborn package
sns.scatterplot(data=melt_cars,x='featureValue',y='price',hue='feature',size='featureValue',palette='flare')

```

```python
sns.scatterplot(data=melt_cars,x='featureValue',y='price',hue='feature',size='featureValue',palette='husl')

```
